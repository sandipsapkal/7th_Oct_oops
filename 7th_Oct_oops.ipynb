```python
Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. 
Implement methods to deposit and withdraw money from the account.
```

```python
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposit of ${amount} successful. New balance is ${self.balance}.")
        else:
            print("Deposit amount should be greater than 0.")

    def withdraw(self, amount):
        if amount > 0:
            if self.balance >= amount:
                self.balance -= amount
                print(f"Withdrawal of ${amount} successful. New balance is ${self.balance}.")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount should be greater than 0.")

# Example usage:
account = BankAccount("123456789", "John Doe", 1000)
print(f"Initial balance: ${account.balance}")
account.deposit(500)
account.withdraw(200)
account.withdraw(1500)  # This will result in an "Insufficient funds" message
```

```python
Problem 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. 
Implement methods to calculate the yearly bonus and display employee details.
```

```python
class Employee:
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary

    def calculate_yearly_bonus(self, bonus_percentage):
        bonus = (bonus_percentage / 100) * self.salary
        return bonus

    def display_details(self):
        print(f"Employee ID: {self.employee_id}")
        print(f"Name: {self.name}")
        print(f"Salary: ${self.salary}")

# Example usage:
employee = Employee("E001", "Alice", 50000)
employee.display_details()
bonus_percentage = 10  # 10% bonus
yearly_bonus = employee.calculate_yearly_bonus(bonus_percentage)
print(f"Yearly bonus: ${yearly_bonus}")
```

```python
Problem 3: Vehicle Rental Create a class representing a vehicle rental system.
Implement methods to rent a vehicle, return a vehicle, and display available vehicles.
```

```python
class VehicleRental:
    def __init__(self):
        self.available_vehicles = {
            "Car": 5,
            "Motorcycle": 3,
            "Truck": 2
        }
        self.rented_vehicles = {
            "Car": 0,
            "Motorcycle": 0,
            "Truck": 0
        }

    def rent_vehicle(self, vehicle_type):
        if self.available_vehicles[vehicle_type] > 0:
            self.available_vehicles[vehicle_type] -= 1
            self.rented_vehicles[vehicle_type] += 1
            print(f"{vehicle_type} rented successfully.")
        else:
            print(f"Sorry, no {vehicle_type}s available for rent.")

    def return_vehicle(self, vehicle_type):
        if self.rented_vehicles[vehicle_type] > 0:
            self.available_vehicles[vehicle_type] += 1
            self.rented_vehicles[vehicle_type] -= 1
            print(f"{vehicle_type} returned successfully.")
        else:
            print(f"No {vehicle_type}s are currently rented.")

    def display_available_vehicles(self):
        print("Available Vehicles:")
        for vehicle_type, count in self.available_vehicles.items():
            print(f"{vehicle_type}: {count}")

# Example usage:
rental_system = VehicleRental()
rental_system.display_available_vehicles()

rental_system.rent_vehicle("Car")
rental_system.rent_vehicle("Motorcycle")
rental_system.rent_vehicle("Truck")
rental_system.rent_vehicle("Car")  # This will indicate no cars are available

rental_system.display_available_vehicles()

rental_system.return_vehicle("Motorcycle")
rental_system.display_available_vehicles()
```

```python
Problem 4: Library Catalog Create classes representing a library and a book. 
Implement methods to add books to the library, borrow books, and display available books.
```

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_borrowed = False

    def borrow_book(self):
        if not self.is_borrowed:
            self.is_borrowed = True
            return True
        else:
            return False

    def return_book(self):
        if self.is_borrowed:
            self.is_borrowed = False
            return True
        else:
            return False

    def __str__(self):
        return f"'{self.title}' by {self.author}"


class LibraryCatalog:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def borrow_book(self, book_title):
        for book in self.books:
            if book.title == book_title:
                if book.borrow_book():
                    print(f"Book '{book_title}' borrowed successfully.")
                else:
                    print(f"Book '{book_title}' is already borrowed.")
                return
        print(f"Book '{book_title}' is not available in the library.")

    def return_book(self, book_title):
        for book in self.books:
            if book.title == book_title:
                if book.return_book():
                    print(f"Book '{book_title}' returned successfully.")
                else:
                    print(f"Book '{book_title}' was not borrowed.")
                return
        print(f"Book '{book_title}' is not in the library's collection.")

    def display_available_books(self):
        if not self.books:
            print("No books available in the library.")
        else:
            print("Available Books:")
            for book in self.books:
                if not book.is_borrowed:
                    print(book)

# Example usage:
library = LibraryCatalog()

book1 = Book("The Great Gatsby", "F. Scott Fitzgerald")
book2 = Book("To Kill a Mockingbird", "Harper Lee")
book3 = Book("1984", "George Orwell")

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

library.display_available_books()

library.borrow_book("The Great Gatsby")
library.borrow_book("1984")
library.borrow_book("The Catcher in the Rye")  # Book not available

library.display_available_books()

library.return_book("The Great Gatsby")
library.return_book("To Kill a Mockingbird")  # Book not borrowed

library.display_available_books()
```

```python
Problem 5: Product Inventory Create classes representing a product and an inventory system. 
Implement methods to add products to the inventory, update product quantity, and display available products.
```

```python
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity

    def update_quantity(self, new_quantity):
        self.quantity = new_quantity

    def __str__(self):
        return f"Product ID: {self.product_id}, Name: {self.name}, Price: ${self.price}, Quantity: {self.quantity}"


class Inventory:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def update_product_quantity(self, product_id, new_quantity):
        for product in self.products:
            if product.product_id == product_id:
                product.update_quantity(new_quantity)
                print(f"Quantity for product '{product.name}' updated to {new_quantity}.")
                return
        print(f"Product with ID '{product_id}' not found in inventory.")

    def display_available_products(self):
        if not self.products:
            print("No products available in the inventory.")
        else:
            print("Available Products:")
            for product in self.products:
                print(product)

# Example usage:
inventory_system = Inventory()

product1 = Product(1, "Laptop", 1200, 10)
product2 = Product(2, "Phone", 800, 15)
product3 = Product(3, "Tablet", 500, 20)

inventory_system.add_product(product1)
inventory_system.add_product(product2)
inventory_system.add_product(product3)

inventory_system.display_available_products()

inventory_system.update_product_quantity(2, 25)
inventory_system.update_product_quantity(4, 30)  # Product not found

inventory_system.display_available_products()
```

```python
Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height.
Implement methods to calculate the area and perimeter of the shape.
```

```python
class Shape:
    def __init__(self, length=0, width=0, height=0):
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        pass

    def calculate_perimeter(self):
        pass


class Rectangle(Shape):
    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)


class Square(Shape):
    def calculate_area(self):
        return self.length ** 2

    def calculate_perimeter(self):
        return 4 * self.length


class Triangle(Shape):
    def calculate_area(self):
        # Using Heron's formula to calculate area of a triangle
        s = (self.length + self.width + self.height) / 2
        return (s * (s - self.length) * (s - self.width) * (s - self.height)) ** 0.5

    def calculate_perimeter(self):
        return self.length + self.width + self.height


# Example usage:
rectangle = Rectangle(5, 10)
print("Rectangle Area:", rectangle.calculate_area())
print("Rectangle Perimeter:", rectangle.calculate_perimeter())

square = Square(7)
print("Square Area:", square.calculate_area())
print("Square Perimeter:", square.calculate_perimeter())

triangle = Triangle(3, 4, 5)
print("Triangle Area:", triangle.calculate_area())
print("Triangle Perimeter:", triangle.calculate_perimeter())
```

```python
Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. 
Implement methods to calculate the average grade and display student details.
```

```python
class Student:
    def __init__(self, student_id, name, grades=[]):
        self.student_id = student_id
        self.name = name
        self.grades = grades

    def add_grade(self, grade):
        self.grades.append(grade)

    def calculate_average_grade(self):
        if not self.grades:
            return 0  # Return 0 if no grades available to avoid division by zero
        return sum(self.grades) / len(self.grades)

    def display_details(self):
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        if self.grades:
            print(f"Grades: {', '.join(map(str, self.grades))}")
            print(f"Average Grade: {self.calculate_average_grade()}")
        else:
            print("No grades available for this student.")

# Example usage:
student1 = Student("S001", "Alice", [80, 75, 90, 85])
student1.display_details()

student2 = Student("S002", "Bob")
student2.add_grade(65)
student2.add_grade(70)
student2.display_details()
```

```python
Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. 
Implement methods to add posts, display posts, and search for posts by keyword.
```

```python
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, post_content):
        self.posts.append(post_content)
        print("Post added successfully.")

    def display_posts(self):
        if not self.posts:
            print("No posts available.")
        else:
            print(f"Posts by {self.username}:")
            for index, post in enumerate(self.posts, start=1):
                print(f"Post {index}: {post}")

    def search_posts_by_keyword(self, keyword):
        found_posts = []
        for post in self.posts:
            if keyword.lower() in post.lower():
                found_posts.append(post)
        if found_posts:
            print(f"Posts containing '{keyword}':")
            for post in found_posts:
                print(post)
        else:
            print(f"No posts found containing '{keyword}'.")

# Example usage:
profile = SocialMediaProfile("JohnDoe")
profile.add_post("Enjoying a beautiful day at the beach!")
profile.add_post("Exploring new hiking trails. Amazing views!")
profile.add_post("Trying out new recipes in the kitchen.")
profile.display_posts()

print("\nSearching for posts containing 'hiking':")
profile.search_posts_by_keyword("hiking")

print("\nSearching for posts containing 'travel':")
profile.search_posts_by_keyword("travel")
```

```python
Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. 
Implement methods to add tasks, mark tasks as completed, and display pending tasks.
```

```python
class ToDoList:
    def __init__(self):
        self.tasks = {}

    def add_task(self, task_name, due_date):
        if task_name not in self.tasks:
            self.tasks[task_name] = {"due_date": due_date, "completed": False}
            print("Task added successfully.")
        else:
            print("Task already exists.")

    def mark_task_completed(self, task_name):
        if task_name in self.tasks:
            self.tasks[task_name]["completed"] = True
            print(f"Task '{task_name}' marked as completed.")
        else:
            print("Task not found.")

    def display_pending_tasks(self):
        pending_tasks = [task for task, details in self.tasks.items() if not details["completed"]]
        if not pending_tasks:
            print("No pending tasks.")
        else:
            print("Pending Tasks:")
            for task in pending_tasks:
                print(f"- {task} (Due: {self.tasks[task]['due_date']})")

# Example usage:
to_do_list = ToDoList()

to_do_list.add_task("Buy groceries", "2023-11-20")
to_do_list.add_task("Finish report", "2023-11-15")
to_do_list.add_task("Call John", "2023-11-18")

to_do_list.display_pending_tasks()

to_do_list.mark_task_completed("Finish report")
to_do_list.display_pending_tasks()
```
